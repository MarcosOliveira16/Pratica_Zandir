```python
# solução 2

# criando dicionario para listar e calcular os 'id_cliente' e suas respectivas compras
lista_comprador = {}
for cliente in vendas['id_cliente']:
    print(f'O {cliente} fez uma compra.')
    if cliente not in lista_comprador:
        lista_comprador[cliente] = 1
    else:
        lista_comprador[cliente] += 1

# pegando o maior valor do dicinario criado para exibição e comparação
maior_valor = max(lista_comprador.values())
print(f'\nO maior valor de compras atual é {maior_valor}.\n')

maior_comprador = {}

# percorre e exibi o(s) maior(es) comprador(es)
for chave, valor in lista_comprador.items():
   if valor == maior_valor:
      print(f'O cliente {chave} fez o maior número de compras, sendo {valor} compras.')
```

# Qual é o motivo mais comum nos atendimentos?

```python
# solução 1
atendimento['motivo'].groupby(atendimento['motivo']).count().sort_values(ascending=False)
```

```python
# solução 2

# criando lista para listar os motivos dos atendimentos
motivo_atendimento = []
for motivo in atendimento['motivo']:
    # pegando o motivo mais repetido
    motivo_atendimento.append(motivo)

# pegando o maior (ou mais repetido) valor da lista
motivo_mais_rep = max(motivo_atendimento)
print(f'\nO maior motivo dos atendimento é "{motivo_mais_rep}".')
```

# 2. Consultas SQL:

## Suponha que você tenha acesso a um banco de dados com uma tabela clientes contendo os campos: id_cliente, nome, cidade, estado, data_cadastro.

## Escreva as seguintes consultas SQL:

# Consulta para retornar o total de clientes cadastrados.

```python
# dado q 'id_cliente' deve ser primary key e não se repete.
SELECT COUNT(*) AS Total_clientes # 'AS' usado para dar apelido
FROM clientes;
```

# Consulta para retornar o número de clientes por cidade.

```python
SELECT cidade, COUNT(*) AS numero_clientes
FROM clientes
GROUP BY cidade; # agrupa por 'cidade'
```

# Consulta para listar os 5 estados com mais clientes.

```python
SELECT estado, COUNT(*) AS numero_clientes
FROM clientes
GROUP BY estado # agrupa por 'estado'
ORDER BY numero_clientes DESC # oderna os 'numero_clientes' de forma decrescente
LIMIT 5; # limita a saída a 5 valores
```

# Gráfico de linha para demonstrar a tendência de vendas nos últimos seis meses.

```python
import matplotlib.pyplot as plt
import numpy as np

fonte = {
    'family': 'DejaVu Sans Mono', # fonte
    'color':  'black', # cor
    'weight': 'bold', # estilo - negrito
    'size': 15, # tamanho
}

# Converter a coluna 'data_venda' para o tipo datetime
vendas['data_venda'] = pd.to_datetime(vendas['data_venda'])

# Extrair o nome do mês
vendas['mes'] = vendas['data_venda'].dt.strftime('%B')

# Agrupar por mês e contar o número de vendas
vendas_por_mes = vendas.groupby('mes').size().reindex([
    'January', 'February', 'March', 'April', 'May', 'June',
    'July', 'August', 'September', 'October', 'November', 'December'
], fill_value=0) # 'fill_value' preenche as entradas onde os dados são ausentes, no caso como apenas o mês de setembro possui entrada
# os demais vão aparecer mesmo sem receber valor algum, permitindo a exibição da 'tendência de vendas' de maneria mais clara

# Definir os eixos x (mês) e y (número de vendas)
x = vendas_por_mes.index
y = vendas_por_mes.values

# alterando resolução para melhor exibição do eixo 'x' (motivos)
plt.figure(figsize=(12, 12))

plt.plot(x, y, marker='o')  # 'maker' adiciona marcadores nos pontos
# 'loc' posição no gráfico, 'x' e 'y' posição mais específica
plt.title('Vendas por mês', loc='left', y=0.96, x=0.01, fontdict=fonte)
plt.xlabel('Mês', fontdict=fonte)
plt.ylabel('Número de vendas', fontdict=fonte)

plt.show()
```

# Data Lake (Teórico):

## Considerando que o volume de dados aumentará significativamente nos próximos anos, a empresa está considerando implementar um Data Lake.

## Escreva um breve parágrafo respondendo às seguintes perguntas:

# O que é um Data Lake e como ele pode ajudar na coleta e organização de dados em uma empresa?

**Data Lake**: é um repositório de armazenamento de dados que permite armazenar grandes volumes de dados brutos em seu formato nativo (maneira crua), seja estruturado, semi-estruturado ou não estruturado.
### Algumas de suas características são:
- não ter sua finalidade (dados) definida;
- altamente acessível e rápido de atualizar;
- não possui otimização (dados cru) e logo é mais difícil de usar.

# Quais vantagens um Data Lake oferece em relação a um banco de dados tradicional?

### Flexibilidade:
- **Date Lake**: por não ter definição de como seram usados, os dados podem ser armazenados de qualquer, ou seja, maior flexibilidade.
- **Banco tradicional**: dados devem ser estruturados e organizados desde o início.

### Escalabilidade:
- **Data Lake**: Permite o armazenamento de uma grande quantidade de dados.
- **Banco tradicional**: Geralmente tem dificuldade em crescer de maneira parecida ao "**Data Lake**", ainda mais quando se trata de um banco relacional (estruturado).

### Custo:
- **Data Lake**: pode ser amarzenado na nuvem ou aplicações específicas para Data Lake.
- **Banco tradicional**: Geralmente requer um bom hardware e compra de licenças específicas.

### Capacidade de análise:
- **Data Lake**: possui uma grande gama de ferramentas de análise, como SQL e de aprendizado de máquinas.
- **Banco tradicional**: geralmente limita-se a SQL e possui dificuldades com dados não estruturados.

### Dados não estruturados:
- **Data Lake**: pode armazenar dados não estruturados como textos, vídeos, imagens e semi estruturados como JSON, XML sem necessidade de conversão.
- **Banco tradicional**: usado geralmente para dados estruturados como tabelas e como já falado, não consegue lidar com dados não estruturados.
